```python
# ==========================================
# Cleaning The Data
# ==========================================

# Check for missing values
print(sales_data.isnull().sum())
```

```python
# Drop rows with any missing values
sales_data_dropped = sales_data.dropna()
```

```python
sales_data_dropped
```

```python
# Drop rows with missing values only in 'Amount' column
sales_data_cleaned = sales_data.dropna(subset=['Amount'])
sales_data_cleaned
```

```python
# Re-check for missing values after cleaning
print(sales_data_cleaned.isnull().sum())
```

```python
# ============================================
#  Slicing and Filtering The Data
# ============================================
# Filter rows where Category = 'Top'
category_data = sales_data[sales_data['Category'] == 'Top']
display(category_data.head())
```

```python
# Filter rows where Amount > 1000
high_amount_data = sales_data[sales_data['Amount'] > 1000]
display(high_amount_data.head())
```

```python
# Filter rows where Category = 'Top' AND Qty = 3
filter_data = sales_data[
    (sales_data['Category'] == 'Top') & (sales_data['Qty'] == 3)
]
display(filter_data.head())
```

```python
# ============================================
#  Aggregating Sales Data
# ============================================

#  Total sales by Category
category_totals = sales_data.groupby('Category', as_index=False)['Amount'].sum()
category_totals = category_totals.sort_values('Amount', ascending=False)
display(category_totals)
```

```python
#  Average sales by Fulfilment & Category
fulfilment_averages = (
    sales_data.groupby(['Category','Fulfilment'], as_index=False)['Amount'].mean()
    .sort_values('Amount', ascending=False)
)
display(fulfilment_averages)
```

```python
#  Average sales by Status & Category
status_averages = (
    sales_data.groupby(['Category','Status'], as_index=False)['Amount'].mean()
    .sort_values('Amount', ascending=False)
)
display(status_averages)
```

```python
#  Total sales by Shipment & Fulfilment
total_sales_shipandfulfil = (
    sales_data.groupby(['Courier Status','Fulfilment'], as_index=False)['Amount'].sum()
    .sort_values('Amount', ascending=False)
    .rename(columns={'Courier Status':'Shipment'})
)
display(total_sales_shipandfulfil)
```

```python
# =============================================================================
# Exporting Data
# =============================================================================
status_averages.to_excel('average_sales_by_Category_and_Status.xlsx', index=False)
total_sales_shipandfulfil.to_excel('total_sales_by_Shipment_and_Fulfilment.xlsx', index=False)
```
